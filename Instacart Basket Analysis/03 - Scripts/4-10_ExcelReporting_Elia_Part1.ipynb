### [3] The Instacart officers are interested in comparing customer behavior in different geographic areas. Create a regional segmentation of the data. You’ll need to create a “Region” column based on the “State” column from your customers data set.
### - Use the region information in this Wikipedia article to create your column (you only need to create regions, not divisions).
### - Determine whether there’s a difference in spending habits between the different U.S. regions. (Hint: You can do this by crossing the variable you just created with the spending flag.)

```python
# Define a function to properly label each Region accordingly to the state column in the dataframe.

def region(row):

  if row['state'] in ('Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'Pennsylvania', 'New Jersey'):
    return 'Northeast'
  elif row['state'] in ('Wisconsin', 'Michigan', 'Illinois', 'Indiana', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri'):
    return 'Midwest'
  elif row['state'] in ('Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana'):
    return 'South'
  elif row['state'] in ('Idaho', 'Montana', 'Wyoming', 'Nevada', 'Utah', 'Colorado', 'Arizona', 'New Mexico', 'Alaska', 'Washington', 'Oregon', 'California', 'Hawaii'):
    return 'West'
  else: return 'N/A'
```

```python
# Apply and create the new column 'Region' with the results of the previous function

df_instacart['region'] = df_instacart.apply(region, axis=1)
```

### [4] The Instacart CFO isn’t interested in customers who don’t generate much revenue for the app. Create an exclusion flag for low-activity customers (customers with less than 5 orders) and exclude them from the data. Make sure you export this sample.

```python
# Then i will use result command to properly label the activity level

result = []

for value in df_instacart["max_order"]:
  if value <= 4:
    result.append("Low Activity")
  elif value >= 5:
    result.append("High Activity")
  else:
    result.append("N/A")
```

```python
# This will assign the results to the new column

df_instacart['customer_activity_level'] = result
```

```python
# Check the dataframe

df_instacart.head(10)
```

```python
# Count the values after creating the new flag/column

df_instacart['customer_activity_level'].value_counts(dropna = False)
```

```python
# Using the shape function to see whats the total if rows and columns before creating the subset

df_instacart.shape
```

```python
# Creating the new dataframe subset, excluding the low activity customers from it

df_instacart_active_customers = df_instacart[df_instacart['customer_activity_level'] == 'High Activity']
```

### [5] The marketing and business strategy units at Instacart want to create more-relevant marketing strategies for different products and are, thus, curious about customer profiling in their database. Create a profiling variable based on age, income, certain goods in the “department_id” column, and number of dependents.
### You might also use the “orders_day_of_week” and “order_hour_of_day” columns if you can think of a way they would impact customer profiles.
### (Hint: As an example, try thinking of what characteristics would lead you to the profile “Single adult” or “Young parent.”)

```python
# Category Young Single Labeling (below 40 years old, no dependants)

df_instacart_active_customers.loc[(df_instacart_active_customers['age'] <= 40) & (df_instacart_active_customers['n_dependants'] == 0), 'customer_profile'] = 'Young Single'
```

```python
# Category Young Parent Shopper Labeling (below 40 years old, at least one dependant)

df_instacart_active_customers.loc[(df_instacart_active_customers['age'] <= 40) & (df_instacart_active_customers['n_dependants'] >= 1), 'customer_profile'] = 'Young Parent Shopper'
```

```python
# Category Old Single Labeling (40 to 65 years old, no dependants)

df_instacart_active_customers.loc[(df_instacart_active_customers['age'] > 40) & (df_instacart_active_customers['age'] <= 65) & (df_instacart_active_customers['n_dependants'] == 0), 'customer_profile'] = 'Old Single'
```

```python
# Category Old Parent Shopper Labeling (from 40 to 65 years old, at least one dependant)

df_instacart_active_customers.loc[(df_instacart_active_customers['age'] > 40) & (df_instacart_active_customers['age'] <= 65) & (df_instacart_active_customers['n_dependants'] >= 1), 'customer_profile'] = 'Old Parent Shopper'
```

```python
# Category Affluent Professional Labeling (25-60 years old, income higher than 100K, no dependants)

df_instacart_active_customers.loc[(df_instacart_active_customers['income'] >= 100000) & (df_instacart_active_customers['n_dependants'] == 0) & (df_instacart_active_customers['age'] >= 25) & (df_instacart_active_customers['age'] <= 60), 'customer_profile'] = 'Affluent Professional'
```

```python
# Category Budget Family Labeling (30-60 years old, income lower than 60K, at least one dependant)

df_instacart_active_customers.loc[(df_instacart_active_customers['income'] <= 60000) & (df_instacart_active_customers['n_dependants'] >= 1) & (df_instacart_active_customers['age'] >= 30) & (df_instacart_active_customers['age'] <= 60), 'customer_profile'] = 'Budget Family'
```

```python
# Category Retired Low Income Shopper (at least 65 years old, income lower than 75K)

df_instacart_active_customers.loc[(df_instacart_active_customers['age'] > 65) & (df_instacart_active_customers['income'] <= 75000), 'customer_profile'] = 'Retired LowIncome Shopper'
```

```python
# Category Retired High Income Shopper (at least 65 years old, income higher than 75K)

df_instacart_active_customers.loc[(df_instacart_active_customers['age'] > 65) & (df_instacart_active_customers['income'] > 75000), 'customer_profile'] = 'Retired HighIncome Shopper'
```

```python
# Check The New Customer Category Counts

df_instacart_active_customers['customer_profile'].value_counts(dropna = False)
```
