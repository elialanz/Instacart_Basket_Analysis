#### 5 - Determine a suitable way to combine the orders_products_combined dataframe with your products data set. Make sure you’re using your wrangled, cleaned, and deduped products data set stored in your “Prepared Data” folder from the previous Exercise’s task.

```python
# Import Dataframe (ProductsChecked.csv)
df_prods_checked = pd.read_csv(os.path.join(path, '02 - Data', 'Prepared Data', 'products_checked.csv'))
```

```python
# I think a normal full join merge will be better, because we want to have all the products to be matched with all the possible orders
# Merge through the product_id column

df_ords_prods_merged = df_ords_prods.merge(df_prods_checked, on = 'product_id', indicator = 'merge_products')
```
