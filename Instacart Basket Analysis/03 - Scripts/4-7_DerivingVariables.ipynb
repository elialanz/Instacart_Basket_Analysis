#### 2 - Suppose your clients have changed their minds about the labels you created in your â€œbusiest_dayâ€ column. Now, they want â€œBusiest dayâ€ to become â€œBusiest daysâ€ (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, theyâ€™d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# First i will rename the column "busiest day" in "busiest days"

ords_prods_merge.rename(columns={'busiest_day': 'busiest_days'}, inplace=True)
```

```python
# Then i will use result command to properly label the two busiest days

result = []

for value in ords_prods_merge["order_day_of_week"]:
  if value == 0:
    result.append("Very Busy")
  elif value == 1:
    result.append("Very Busy")
  else:
    result.append("Not Busy")
```

```python
result
```

```python
# This will assign the results to the column

ords_prods_merge['busiest_days'] = result
```

```python
# Here i will perform a counting check

ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
ords_prods_merge
```

```python
# This is the same approach, but focused on the slowest days

result = []

for value in ords_prods_merge["order_day_of_week"]:
  if value == 4:
    result.append("Very Slow")
  elif value == 3:
    result.append("Very Slow")
  else:
    result.append("Not Slow")
```

#### 4 - When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled â€œMost orders,â€ â€œAverage orders,â€ and â€œFewest orders.â€ Create a new column containing these labels called â€œbusiest_period_of_day.â€

```python
# First i will count the frequency of orders/hours, and divide them in 3 groups of 8, from top to bottom

ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Then i will use result command to properly label the three groups

result = []

for value in ords_prods_merge["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result.append("Most Orders")
    elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
        result.append("Average orders")
    else:
        result.append("Fewest orders")
```

```python
result
```

```python
ords_prods_merge['busiest_period_of_day'] = result
```
